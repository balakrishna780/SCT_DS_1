```python
import pandas as pd
import matplotlib.pyplot as plt

# Create dataset
data = {
    'Age Group': ['0-20 Years', '21-64 Years', '65+ Years'],
    'Population (Millions)': [512, 807, 98],
    'Percentage (%)': [36.1, 57.0, 6.9]
}

df = pd.DataFrame(data)

# Plot bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(df['Age Group'], df['Population (Millions)'], color=['yellow', 'blue', 'pink'], edgecolor='black')

# Add percentages on top
for bar, pct in zip(bars, df['Percentage (%)']):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, height + 10, f'{pct}%', ha='center', fontsize=12)

# Add labels and title
plt.title("India's Population Distribution by Age in 2022", fontsize=14)
plt.xlabel("Age Groups")
plt.ylabel("Population (Millions)")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
plt.savefig("India_Population_Age_Distribution_2022.png")
```


**Objective:**  
Visualize the age-wise population of India using a bar chart.
